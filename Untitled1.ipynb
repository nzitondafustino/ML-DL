```python
class  MyDataSet(Dataset):
  def __init__(self, data_path, label_path, context):
    self.X = np.load(data_path,allow_pickle=True)
    self.Y = np.load(label_path,allow_pickle=True)
    # self.X = X
    # self.Y = Y
    self.indeces = []
    for i,x in enumerate(self.X):
      for j,xx in enumerate(x):
        self.indeces.append((i,j))
    self.l = len(self.indeces)
    
    for i,x in enumerate(self.X):
            self.X[i] = np.pad(x,((context,context),(0,0)),'constant',constant_values=0)

  def __len__(self):
    return self.l

  def __getitem__(self,index):
    i,j = self.indeces[index]
    return self.X[i][j,:],self.Y[i][j]

  def collate_fn(batch):
    X = [x for (x,y) in batch]
    Y = [y for (x,y) in batch]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    return X,Y
    # print(batch)
```

```python
train_dataset = MyDataSet(data_path,labels_path,10)
train_dataloader = DataLoader(train_dataset, batch_size=2,shuffle=True, collate_fn=MyDataSet.collate_fn)
```

```python
class NeuralNet(nn.Module):
  def __init__(self,input_feature,hidden_layers,out_features):
    super().__init__()
    self.layers = []
    prevous_features = input_feature
    for i,l in enumerate(hidden_layers):
      linear = nn.Linear(prevous_features,l)
      self.layers.append(linear)
      self.layers.append(nn.ReLU())
      self.layers.append(nn.BatchNorm1d(l))
      prevous_features = l
    self.layers.append(nn.Linear(hidden_layers[-1],out_features))
    self.fpass = nn.Sequential(*self.layers)
  def forward(self,x):
    return self.fpass(x)
```

```python
neuralNet = NeuralNet(100,[50,40,20,15],10)
print(neuralNet)
```

```python
ccriterion = nn.CrossEntropyLoss()
optimizer = torch.nn.optimizer(neuralNet.parameters(),lr=0.)
```
