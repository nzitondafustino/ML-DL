```python
class My_range:
  def __init__(self,start,end):
    self.start = start
    self.end = end
  def __iter__(self):
    return self
  def __next__(self):
    current = self.start
    if self.start >= self.end:
      raise StopIteration
    else:
      self.start +=1
      return current
```
