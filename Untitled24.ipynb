```python
# img =cv2.imread("/content/testImages/" + imagesList[2])
img =cv2.imread("/content/testImages/1828_1942-10-11_2013.jpg")
img.shape
```

```python
haar_detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
def detect_faces(img):
   gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   faces = haar_detector.detectMultiScale(gray, 1.3, 5)
   return faces
 
faces = detect_faces(img)
for x, y, w, h in faces:
   detected_face = img[int(y):int(y+h), int(x):int(x+w)]
```

```python
if not faces:
  haar_detector = cv2.CascadeClassifier("haarcascade_profileface.xml")
  faces = detect_faces(img)
  for x, y, w, h in faces:
    detected_face = img[int(y):int(y+h), int(x):int(x+w)]
  if not faces:
    haar_detector = cv2.CascadeClassifier("haarcascade_upperbody.xml")
    faces = detect_faces(img)
    for x, y, w, h in faces:
      detected_face = img[int(y):int(y+h), int(x):int(x+w)]
    if not faces:
      detected_face = img
```

```python
detected_face = cv2.resize( detected_face, (224, 224))
blob_image = cv2.dnn.blobFromImage(detected_face)
```
