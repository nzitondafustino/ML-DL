```python
class My_range:
  def __init__(self,start,stop):
    self.start = start
    self.stop = stop
  def __next__(self):
    current = self.start
    if self.start >= self.stop:
      raise StopIteration
    else:
      self.start +=1
      return current
  def __iter__(self):
    return self
  
def my_gen_range(start,stop):
  current = start
  while current < stop:
    yield current
    current += 1
```

```python
for i in My_range(0,20):
  print(i)
print()

for i in my_gen_range(0,20):
  print(i)
```
