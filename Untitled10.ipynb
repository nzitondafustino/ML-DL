```python
import numpy as np
def dilateAndPad1d(arr,pady=2,padx=2,stride=1):
    z,y,x = arr.shape
    mid_point = x//2
    b = np.zeros((z,y+pady*2,x+padx*2 + stride-1))
    y_end = y+pady*2
    x_end = x+padx*2 + stride-1
    b[:,pady:y_end-pady,padx:padx+mid_point]= arr[:,:,:mid_point]
    b[:,pady:y_end-pady,-(padx+mid_point):x_end-padx] = arr[:,:,-mid_point:]
    return b
```
